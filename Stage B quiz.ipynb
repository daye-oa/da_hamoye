Processing Data 


```python
sc_l = MinMaxScaler()
```

```python
df_1 = df.drop(columns = ['date' , 'lights'])
df_1.head()
```

```python

df_t =pd.DataFrame(sc_l.fit_transform(df_1), columns = df_1.columns)


X_df = df_1.drop(columns = ['Appliances'] )

y_df = df_1['Appliances']

```

```python
# Splitting trainin data and test data
x_train, x_test, y_train, y_test = train_test_split(X_df, y_df, test_size=0.3, random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
print (y_train.shape)
print(y_test. shape)

```

```python

lir = LinearRegression()

lir.fit(x_train, y_train)
pred_y = lir.predict(x_test)
```

```python
#Mean absolute error 

mae =  metrics.mean_absolute_error(y_test, pred_y)
print(mae)
```

```python
#Residula sum of sqaures

rss = np.sum(np.square(y_test - pred_y))
print (rss)
```

```python
#root mean squared error 

mse = metrics.mean_squared_error(y_test, pred_y)
print (np.sqrt(mse))
```

```python
# r2 score 

r_2 = metrics.r2_score(y_test, pred_y)
print (r_2)
```

```python
# the ridge model on the train, test data

rid_reg = Ridge(alpha = 0.4)

rid_reg.fit(x_train, y_train)
rid_y = rid_reg.predict(x_test)

mse = metrics.mean_squared_error(y_test, rid_y)
print (np.sqrt(mse))
```

```python
# root mean square error of the Lasso model 

lass_reg = Lasso(alpha = 0.001)
lass_reg.fit(x_train, y_train)

lass_y = lass_reg.predict(x_test)
mse = metrics.mean_squared_error(y_test, lass_y)
print (np.sqrt(mse))
```

```python
def get_weights_df(model, feat, col_name):
  #this function returns the weight of every feature
  weights = pd.Series(model.coef_, feat.columns).sort_values()
  weights_df = pd.DataFrame(weights).reset_index()
  weights_df.columns = ['Features', col_name]
  weights_df[col_name].round(3)

  return weights_df
```

```python
lasso_weights_df = get_weights_df(lass_reg, x_train, 'Lasso_weight')
lasso_weights_df
```
